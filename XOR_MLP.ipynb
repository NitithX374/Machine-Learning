```python
import numpy as np

x1x2 = [[0, 0], [0, 1], [1, 0], [1, 1]]
y = [0, 1, 1, 0]

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def forward(x1, x2):
    w1 = 5
    w2 = 10
    b = -15
    z = w1 * x1 + w2 * x2 + b
    return sigmoid(z)
def backward(x1, x2, y_true):
    w1 = 10
    w2 = 10
    b = -15
    z = w1 * x1 + w2 * x2 + b
    y_pred = sigmoid(z)
    
    # Calculate the error
    error = y_pred - y_true
    
    # Calculate gradients
    d_y_pred = error
    d_z = d_y_pred * sigmoid(z) * (1 - sigmoid(z))
    
    d_w1 = d_z * x1
    d_w2 = d_z * x2
    d_b = d_z
    
    return d_w1, d_w2, d_b
def update_parameters(w1, w2, b, d_w1, d_w2, d_b, learning_rate):
    w1 -= learning_rate * d_w1
    w2 -= learning_rate * d_w2
    b -= learning_rate * d_b
    return w1, w2, b
def train(x1x2, y, epochs, learning_rate):
    w1 = 10
    w2 = 10
    b = -15
    
    for epoch in range(epochs):
        for (x1, x2), y_true in zip(x1x2, y):
            # Forward pass
            y_pred = forward(x1, x2)
            # Backward pass
            d_w1, d_w2, d_b = backward(x1, x2, y_true)
            # Update parameters
            w1, w2, b = update_parameters(w1, w2, b, d_w1, d_w2, d_b, learning_rate)
    return w1, w2, b

# Train the model
epochs = 10000
learning_rate = 0.001
w1, w2, b = train(x1x2, y, epochs, learning_rate)
print(f"Trained parameters: w1={w1}, w2={w2}, b={b}")
# Test the model
for (x1, x2), y_true in zip(x1x2, y):
    y_pred = forward(x1, x2)
    print(f"Input: ({x1}, {x2}), Predicted: {y_pred:.4f}, True: {y_true}")
```
